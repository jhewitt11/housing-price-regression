```python
import pandas as pd
import numpy as np

from scipy.stats import skew
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

pd.options.display.max_columns = None
pd.options.display.max_rows = 100
```

```python
train_df.head()
```

```python
train_df.describe(include = 'all')
```

### Peel Id and SalePrice columns

```python
y = train_df[['Id', 'SalePrice']]
train_df =  train_df.drop(columns = ['Id', 'SalePrice'])
test_df = test_df.drop(columns = ['Id'])
```

```python
train_df = pd.concat([train_df, test_df], axis = 0)
```

## Recast 'MSSubClass'

```python
train_df['MSSubClass'] = train_df['MSSubClass'].astype('object')
```

# Fill missing values

### for numeric - use mean
### for categorical  use a) mode b) 'None'

```python
columns_NA_meaningful = [
    'Alley',
    'BsmtQual',
    'BsmtCond',
    'BsmtExposure',
    'BsmtFinType1',
    'BsmtFinType2',
    'FireplaceQu',
    'GarageType',
    'GarageFinish',
    'GarageQual',
    'GarageCond',
    'PoolQC',
    'Fence',
    'MiscFeature'   
]
```

```python
for col in train_df.loc[:, train_df.isna().sum() > 0].columns :

    if train_df[col].dtype == 'object':   
        
        # columns where NA is meaningful
        if col in columns_NA_meaningful : 
            train_df[col] = train_df[col].fillna('None')
        
        else :
            train_df[col] = train_df[col].fillna(train_df[col].mode()[0])
        
    else :
        train_df[col] = train_df[col].fillna(train_df[col].mean())
```

```python
train_df.isna().sum()
```

```python
train2 = train_df.copy()
```

# Feature  Transformation

#### input  features
#### target scaling

```python
train2.head()
```

```python
skew(train2['MoSold'].values)
```

## Numeric types

#### If py.sci skew > 0.5, correct with np.log1p

```python
for col in train2.columns:
    
    if train2[col].dtypes != 'object':
        skewness = np.abs(skew(train2[col].values))
        
        if skewness >= 0.5 :
            train2[col] = np.log1p(train2[col])
```

#### Months in a year is cyclical -> use cosine to translate that

```python
train2['MoSold'].unique()
```

```python
train2['MoSold'] = train2['MoSold'].apply(lambda x : -np.cos(0.5236 * x))
```

```python
train2.head()
```

#### Standard Scaler + 1hot encoding

```python
# On non-object columns
numerics = train2.select_dtypes(exclude = 'object')
categoricals = train2.select_dtypes('object')


scaler = StandardScaler()
scaler.fit(numerics)

numerics = pd.DataFrame(scaler.transform(numerics), columns = numerics.columns)
categoricals = pd.get_dummies(categoricals, dtype = 'float')
```

```python
numerics.head()
```

```python
categoricals.head()
```

```python
numerics.reset_index(drop = True, inplace=True)
categoricals.reset_index(drop = True, inplace=True)
```

```python
train3 = pd.concat([numerics, categoricals], axis = 1)
```

```python
train3.head()
```
